```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_excel("WindEnergyForecastingData.xlsx")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
timestamp = list(df['TIMESTAMP'])
time = []
for i in range(0,len(df['TIMESTAMP'])):
    l,r = list(map(str, timestamp[i].split(" ")))
    time.append(r)
```

```python
print(time)
```

```python
timestamp = []
for i in range(0,len(time)):
    l,r = list(map(str, time[i].split(":")))
    timestamp.append(l)
```

```python
print(timestamp)
```

```python
df['TIMESTAMP'] = timestamp
```

```python
print(df['TIMESTAMP'])
```

```python
df.head(5)
```

```python
df['TIMESTAMP'] = df['TIMESTAMP'].astype(float)
```

```python
df.dtypes
```

```python
y = df['TARGETVAR']
X = df.drop(['TARGETVAR'],axis=1)
```

```python
print(y.head(5))
```

```python
print(X.head(5))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=17)
```

```python
from sklearn.svm import SVR
model = SVR(gamma='auto')
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
import matplotlib.pyplot as plt
plt.plot(y_test,y_pred,marker='.',linestyle='none')
plt.title("Prediction") 
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.tight_layout()
```
